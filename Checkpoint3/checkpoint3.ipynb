```python
# Jetez un oeil à cette fonction
# Essayez d'exécuter la fonction, analysez ce qu'elle fait exactement et ses utilités, 
# écrivez un paragraphe décrivant votre analyse (le paragraphe doit être en anglais).

def plot_correlation_map(df):
    
    # Calculer la matrice de corrélation
    corr = df.corr()
    # Configurer la figure matplotlib
    s,ax = plt.subplots(figsize=(12 , 10))
    # Générer une palette de couleurs divergentes personnalisée
    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True)
    # Dessinez la carte thermique avec le masque et corrigez le format d'image
    s = sns.heatmap(corr, cmap = cmap, square=True, cbar_kws={ 'shrink' : .9 }, ax=ax, annot = True, annot_kws = { 'fontsize' : 12 })
    
    
    
plot_correlation_map(df)


# analyse en francais:
# Cette fonction permet de calculer la matrice de corrélation
# tout en generant une palette de couleurs divergentes et personnalisée.
# Cette fonction permet de dessiner la carte thermique avec le masque et corriger le format d'image.

# analyse in english:
# This function calculates the correlation matrix
# while generating a divergent and personalized color palette.
# This function draws the heat map with the mask and corrects the aspect ratio.
```
